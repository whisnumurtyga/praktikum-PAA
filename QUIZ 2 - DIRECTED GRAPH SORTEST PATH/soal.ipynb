```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


def draw_directed_graph(matrix, nodes, positions):
    G = nx.DiGraph()

    # Tambahkan simpul ke graf
    for node in nodes:
        G.add_node(node)

    # Tambahkan tepi ke graf berdasarkan matriks ketetanggaan
    for i in range(len(nodes)):
        for j in range(len(nodes)):
            if matrix[i][j] > 0:
                G.add_edge(nodes[i], nodes[j], weight=matrix[i][j])

    # Tentukan posisi simpul berdasarkan sub posisi yang diberikan
    pos = {}
    for node in nodes:
        if node in positions[0]:
            pos[node] = (1, 5)
        elif node in positions[1]:
            pos[node] = (2, 4 - positions[1].index(node))
        elif node in positions[2]:
            pos[node] = (3, 3 - positions[2].index(node))
        elif node in positions[3]:
            pos[node] = (4, 2 - positions[3].index(node))
        elif node in positions[4]:
            pos[node] = (5, 1)

    # Gambar directed graph di satu figur terpisah
    plt.figure(1)
    nx.draw_networkx(G, pos, node_size=500, node_color='lightblue', font_size=12, font_weight='bold', with_labels=True, arrowsize=20, width=1.5)
    labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
    plt.axis('off')


def show_image(image_path):
    # Membaca gambar dari path yang diberikan
    img = mpimg.imread(image_path)

    # Menampilkan gambar menggunakan imshow
    plt.figure(2)
    plt.imshow(img)
    plt.axis('off')  # Tidak menampilkan sumbu koordinat

    # Menampilkan gambar
    plt.show()
```
